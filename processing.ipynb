```python
import re

def process_value(value):
    # 문자열인지 확인 (숫자나 NaN 처리 대비)
    if isinstance(value, str):
        # 숫자로 시작하는지 확인하고, 숫자 제거
        new_value = re.sub(r'^\d+', '', value)
        
        # 첫 번째 알파벳을 대문자로 변경
        if new_value and new_value[0].isalpha():
            new_value = new_value[0].upper() + new_value[1:]
        return new_value
    else:
        return value

def process_dataframe(df, column_name):
    df[column_name] = df[column_name].apply(process_value)
    return df


# notes dataframe 전처리 
notes_df = process_dataframe(notes_df, 'note')


# 향수 dataframe note column 전처리 
for i in range(1, 21):
    data = process_dataframe(data, f'note{i}')
```

```python
def convert_to_lower(text):

    words = text.split()
    
    converted_words = [words[0][0].upper() + words[0][1:]]
    converted_words += [word[0].lower() + word[1:] if word[0].isupper() else word for word in words[1:]]

    return ' '.join(converted_words)
```

```python
filtering = {
    'women' : {
    'casual&daily' : [['F', 'U'], ['citrus', 'green', 'white_floral']],
    'lovely&femin' : [['F'], ['floral', 'white_floral', 'animal']],    
    'chic&cool' : [['F', 'U'], ['woody', 'green', 'synthetic']],
    'vintage&street' : [['F', 'U'], ['spicy', 'animal', 'balsamic']],
    'kitch&cute' : [['F'], ['fruit', 'citrus', 'sweet']]      
    },
    
    'men' : {       
    'casual&daily' : [['M', 'U'], ['citrus', 'green', 'woody']],
    'dandy' : [['M'], ['woody', 'green' ,'white_floral']],    
    'formal&chic' : [['M', 'U'], ['woody', 'spicy', 'synthetic']],
    'vintage&street' : [['M', 'U'], ['spicy', 'animal', 'balsamic']],
    'sporty' : [['M'], ['animal', 'green', 'citrus']]      
    }
}
```

```python
##### user input
gender = 'women'
style_1 = 'vintage&street'
style_2 = 'chic&cool'
style_3 = None

#### filtering fuction
def filtering_func(gender, style_1, style_2, style_3):
    # 선택한 성별과 스타일에 맞는 향수 필터링 
    
    gender_list = []
    group_list = []
    
    for style in [style_1, style_2, style_3]:
        if style == None:
            break
        else:     
            gender_list.extend(filtering[gender][style][0])
            group_list.extend(filtering[gender][style][1])
    
    return list(set(gender_list)), list(set(group_list))

gender_fil, group_fil = filtering_func(gender, style_1, style_2, style_3)
```

```python
# 성별로 필터링
gender_filter = data_by_groups[data_by_groups['gender'].isin(gender_fil)]

# 그룹으로 필터링 
group_cols = [f'group{i}' for i in range(1, 21)]
group_filter = gender_filter[gender_filter[group_cols].isin(group_fil).any(axis=1)]
```
